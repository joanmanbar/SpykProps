# Batch Execution

```python
def SpykBatch(Images,TROE2020=False,Path_ImgOut=None):
    
    import time
    from datetime import datetime, date

    start_time = time.time()
    
    Nimages = str(len(Images))
    
    Spikes_data = pd.DataFrame()
    Annot_data = pd.DataFrame()
    Contours_data = pd.DataFrame()
    Spklts_data = pd.DataFrame()
    Distances_data = pd.DataFrame()
    Contours_data_annot = pd.DataFrame()
    Spklts_data_annot = pd.DataFrame()
    Distances_data_annot = pd.DataFrame()
    
    Counter=0


    for img_name in Images:
        
        try:

            Counter = Counter+1
            Progress = str(Counter) + "/" + Nimages
            # img_name=Images[0]
            now = datetime.now().strftime("%H:%M:%S")
            print("Processing image " + Progress + ": ", img_name + " at " + now)
            # img_name = Images[9]

            # Set the initial time per image
            image_time = time.time()

            # Remove background and create images
            I = RemoveBackground(img_name, OtsuScaling=0.25, 
                                 rgb_out=True, gray_out=True, 
                                 lab_out=True, hsv_out=True, 
                                 bw_out=True)
            rgb0 = I[0]
            gray0 = I[1]
            lab0 = I[2]
            hsv0 = I[3]
            bw0 = I[4]
            
            # Add annotation
            bw_annot = plt.imread(Masks[Counter-1])
            bw_annot = bw_annot.astype(np.uint8)
            bw_annot = bw_annot[:,:,0]
            bw_annot = bw_annot[44:6940, 25:4970]
            I2 = I[:4].copy()
            I2.append(bw_annot.copy())
            
            
            # Enumerate spikes
            NameOut_spk = Path_ImgOut+'EnumeratedCV/'+img_name.split('\\')[-1]+'.png' 
            EnumerImg_spk = EnumerateSpkCV(bw0, rgb0, 
                                           TextSize=5, TROE2020=False, Plot=False, PlotOut=True)
            # Save image 
            EnumerImg_spk = Image.fromarray(EnumerImg_spk)
            EnumerImg_spk.save(NameOut_spk)
            
            # Enumerate annotations
            NameOut_spk = Path_ImgOut+'EnumeratedCV_annot/'+img_name.split('\\')[-1]+'.png' 
            EnumerImg_spk = EnumerateSpkCV(bw_annot, rgb0, 
                                              TextSize=2, TROE2020=False, Plot=False, PlotOut=True)
            # Save image 
            EnumerImg_spk = Image.fromarray(EnumerImg_spk)
            EnumerImg_spk.save(NameOut_spk)

            # Collect spikes data
            df = SpikesDF(I=I, ImagePath=img_name, RemoveBG=False, rm_envelope=False)
            df_annot = SpikesDF(I=I2, ImagePath=img_name, RemoveBG=False, rm_envelope=False)

            # Label spikes
            labeled_spks, num_spikes = label(bw0, return_num = True)
            # plt.imshow(labeled_spks==1)

            # Spike lengths
            SpkLengths = []

            # Spike distances from each other
            SpkDists = []
            SpkDists_annot = []

            # TROE2021 and PGR start at 2. Ignore background (0) and envelope (1)
            # if TROE2020==False:
            #     FirstSpk = 2
            # else:
            #     FirstSpk = 1

            ContoursPerSpk = pd.DataFrame()
            SpkltsPerSpk = pd.DataFrame()
            DistancesPerSpk = pd.DataFrame()    
            ContoursPerSpk_annot = pd.DataFrame()
            SpkltsPerSpk_annot = pd.DataFrame()
            DistancesPerSpk_annot = pd.DataFrame()

            # Start at 1 so it ignores background (0) 
            for Label in range(1, num_spikes+1):
                
                try:
                    print("Processing spike ", Label)
            #         plt.imshow(labeled_spks==2)

                    spk = labeled_spks==Label
                    # io.imshow(spk)

                    # Crop spike
                    slice_x, slice_y = ndimage.find_objects(spk)[0]
                    cropped_spk = spk[slice_x, slice_y]
                    cropped_annot = bw_annot[slice_x, slice_y]
                    cropped_annot = cropped_annot*cropped_spk
                    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
                    cropped_annot = cv2.morphologyEx(cropped_annot, cv2.MORPH_CLOSE, kernel)
                    cropped_annot = ndimage.binary_fill_holes(cropped_annot).astype(int)
                    cropped_rgb = rgb0[slice_x, slice_y]
                    cropped_rgb_annot = np.where(cropped_annot[..., None], cropped_rgb, 0)
                    cropped_rgb = np.pad(cropped_rgb, pad_width=[(100, 100),(100, 100),(0, 0)], mode='constant')
                    cropped_rgb_annot = np.pad(cropped_rgb_annot, pad_width=[(100, 100),(100, 100),(0, 0)], mode='constant')
                    cropped_spk = np.pad(cropped_spk, pad_width=[(100, 100),(100, 100)], mode='constant')
                    cropped_annot = np.pad(cropped_annot, pad_width=[(100, 100),(100, 100)], mode='constant')
                    
                    cropped_rgb = np.where(cropped_spk[..., None], cropped_rgb, 0)
                    cropped_gray = color.rgb2gray(cropped_rgb)
                    cropped_lab = color.rgb2lab(cropped_rgb)
                    cropped_hsv = color.rgb2hsv(cropped_rgb)
            #         io.imshow(cropped_spk)

                    cropped_rgb_annot = np.where(cropped_annot[..., None], cropped_rgb, 0)
                    cropped_gray_annot = color.rgb2gray(cropped_rgb_annot)
                    cropped_lab_annot = color.rgb2lab(cropped_rgb_annot)
                    cropped_hsv_annot = color.rgb2hsv(cropped_rgb_annot)
                    # plt.imshow(cropped_rgb_annot)


                    # Spike length
                    NameOut_L = Path_ImgOut+'spk_length/'+img_name.split('\\')[-1]+'_spk'+str(Label)+'.png'
                    SL = spk_length(cropped_spk, Overlay=False, Save=True, Filename=NameOut_L)
                    SpkLengths.append(SL)
                    
                    print("Got Spikel length")
                    
                    # Theshold for detection
                    NameOut_thr = Path_ImgOut+'spklt_thr/'+img_name.split('\\')[-1]+'_spk'+str(Label)+'.png'
                    thresh2 = SpkltThresh(cropped=cropped_rgb, ResizeFactor=30, 
                                          thr2=0.8, MinSize=1000, Save=True, Filename=NameOut_thr)
                    
                    print("Got Spikel thre")
                    
                    # Remove padding
                    # thresh2 = thresh2[100:-100, 100:-100]
                    # plt.imshow(thresh2)                    
                    
                    # Distances between contours
                    D = DistAll(bw=thresh2, HeatMap=False, spike_length=SL)
                    D_annot = DistAll(bw=cropped_annot, HeatMap=False, spike_length=SL)
                    SpkDists.append(D)
                    SpkDists_annot.append(D_annot)
                    # Image_Name = [img_name.split('\\')[-1]]
                    # DistMat = pd.DataFrame(list(zip(Image_Name, str(Label), D)), columns = ['Image_Name', 'Spike_Label', 'MatrixD'])
                    
                    print("Got distances")
                    
                    # Contours                   
                    NameOut_cont = Path_ImgOut+'contours/'+img_name.split('\\')[-1]+'_spk'+str(Label)+'.png'
                    labels_cont, OutImage_cont = CountorProps(cropped_rgb, cropped_lab, cropped_hsv, 
                                               thresh2=thresh2, ImagePath=img_name, 
                                               ResizeFactor=30, MinSize=1000, thr2=0.8, 
                                               Plot=True, Save=True)
                    # Save image 
                    OutImage_cont = Image.fromarray(OutImage_cont)
                    OutImage_cont.save(NameOut_cont)
                                        
                    print("Should have written contours")
                    
                    NameOut_cont = Path_ImgOut+'contours/annot_'+img_name.split('\\')[-1]+'_spk'+str(Label)+'.png'
                    labels_cont_annot, OutImage_cont = CountorProps(cropped_rgb_annot, cropped_lab_annot, cropped_hsv_annot, 
                                               thresh2=np.uint8(cropped_annot), ImagePath=img_name, 
                                               ResizeFactor=30, MinSize=1000, thr2=0.8, 
                                               Plot=True, Save=True)
                    # Save image 
                    OutImage_cont = Image.fromarray(OutImage_cont)
                    OutImage_cont.save(NameOut_cont)
                                        
                    print("Should have written contours annot")
                    
                    # Add spike label
                    labels_cont['Spike_Label'] = [str(Label)] * len(labels_cont)
                    labels_cont_annot['Spike_Label'] = [str(Label)] * len(labels_cont_annot)

                    # Watershed segments
                    NameOut = Path_ImgOut+'watershed/'+img_name.split('\\')[-1]+'_spk'+str(Label)+'.png'
                    spklts, n_spklts, Angles, OutImage = LabelSpklts(cropped_rgb, MinDist=50, labels_out=True,
                              n_spklt=True, ElliPlot=True, Plot=True, Save=True)
                    # Save image 
                    OutImage = Image.fromarray(OutImage)
                    OutImage.save(NameOut)
                    
                    NameOut = Path_ImgOut+'watershed/annot_'+img_name.split('\\')[-1]+'_spk'+str(Label)+'.png'
                    spklts_annot, n_spklts_annot, Angles_annot, OutImage_annot = LabelSpklts(cropped_rgb_annot, MinDist=50, labels_out=True,
                              n_spklt=True, ElliPlot=True, Plot=True, Save=True)
                    # Save image 
                    OutImage = Image.fromarray(OutImage_annot)
                    OutImage.save(NameOut)
                    
                    # Remove padding
                    # spklts = spklts[100:-100, 100:-100]
                    # spklts_annot = spklts_annot[100:-100, 100:-100]
                    
                    WSProps = ObjProps(labeled=spklts, cropped_rgb=cropped_rgb, cropped_lab=cropped_lab,
                                       cropped_hsv=cropped_hsv, ImagePath=img_name, MinSize = 5000)

                    WSProps_annot = ObjProps(labeled=spklts_annot, cropped_rgb=cropped_rgb_annot, cropped_lab=cropped_lab_annot,
                                       cropped_hsv=cropped_hsv_annot, ImagePath=img_name, MinSize = 5000)
                    
                    # Add spike label
                    WSProps['Spike_Label'] = [str(Label)] * len(WSProps)
                    WSProps_annot['Spike_Label'] = [str(Label)] * len(WSProps_annot)

                    ContoursPerSpk = ContoursPerSpk.append(labels_cont)
                    ContoursPerSpk_annot = ContoursPerSpk_annot.append(labels_cont_annot)
                    SpkltsPerSpk = SpkltsPerSpk.append(WSProps)
                    SpkltsPerSpk_annot = SpkltsPerSpk_annot.append(WSProps_annot)
                   
                    # DistancesPerSpk = DistancesPerSpk.append(DistMat)
                    
                except:
                    pass

            Contours_data = Contours_data.append(ContoursPerSpk)
            Contours_data_annot = Contours_data_annot.append(ContoursPerSpk_annot)
            Spklts_data = Spklts_data.append(SpkltsPerSpk)
            Spklts_data_annot = Spklts_data_annot.append(SpkltsPerSpk_annot)
            # Distances_data = Distances_data.append(SpkDists)


            # Add spike lengths and append current data frame
            df["SpykLength"] = SpkLengths
            Spikes_data = Spikes_data.append(df)
            Annot_data = Annot_data.append(df_annot)

            # Distances matrix

            # Create columns with image name, spike index, and distances matrix
            Image_Name = img_name.split('\\')[-1]
            Image_Name = [Image_Name] * len(SpkDists)
            Image_Name_annot = [Image_Name] * len(SpkDists_annot)
            Spk_Index = [number for number in range(1, num_spikes+1)]
            # Matrix of distances among detected spikelets
            DistMat = pd.DataFrame(list(zip(Image_Name, Spk_Index, SpkDists)), columns = ['Image_Name', 'Spike_Label', 'MatrixD'])
            DistMat_annot = pd.DataFrame(list(zip(Image_Name_annot, Spk_Index, SpkDists_annot)), columns = ['Image_Name', 'Spike_Label', 'MatrixD'])
            # Append
            Distances_data = Distances_data.append(DistMat)
            Distances_data_annot = Distances_data_annot.append(DistMat_annot)

            # How long did it take to run this image?
            print("Image " + img_name.split('\\')[-1] + ", " + Progress + ", was fully processed in " + 
                  str(round(time.time() - image_time, 1)) + " seconds. " + "\n")
            
        except:
            pass

    # How long did it take to run the whole code?
    print("This entire code took", str(round(time.time() - start_time, 1)), "seconds to run.")
    
    return Spikes_data,Annot_data,Contours_data,Spklts_data,Distances_data,Contours_data_annot,Spklts_data_annot,Distances_data_annot
```

```python
def EnumBatch(Images, TROE2020=False, TextSize=3, OtsuScaling=0.25):
    
    import time
    start_time = time.time()
    
    Nimages = str(len(Images))
    Counter=0
    
    for img_name in Images:

        Counter = Counter+1
        Progress = str(Counter) + "/" + Nimages
        # img_name=Images[0]
        print("Enumerating image " + Progress + ": ", img_name)

        # Set the initial time per image
        image_time = time.time()

        # Remove background and create images
        I = RemoveBackground(img_name, OtsuScaling=OtsuScaling, rgb_out=True, gray_out=False, lab_out=False, hsv_out=False, bw_out=True)
        rgb0 = I[0]
        bw0 = I[1]

        EnumPlot = EnumerateSpkCV(bw0, rgb0, TextSize=TextSize, TROE2020=False, Plot=False, PlotOut=True)
        ThisImage = img_name.split('\\')[-1:]
        plotname = '.\Images\ENUMERATED\Validation\\' + ThisImage[0] + '.jpg'
        plt.imsave(plotname, EnumPlot)           

        # How long did it take to run this image?
        print("Image " + img_name.split('\\')[-1] + ", " + Progress + ", was enumerated in " + str(round(time.time() - image_time, 1)) + " seconds." + "\n")
            
            
    # How long did it take to run the whole code?
    print("This entire code took", str(round(time.time() - start_time, 1)), "seconds to run.")
    
```

```python
# Detect green spikes
GList = TR_spk["Green"].str.split(",", n = -1, expand = False)
Gdf = pd.DataFrame(GList)
# Gdf.columns = ['spk_0', 'spk_1', 'spk_2', 'spk_3']
Gdf = pd.DataFrame(Gdf.Green.values.tolist(), Gdf.index).add_prefix('spk_')
Gdf2 = pd.concat([TR_spk[['Image_Name','Green']], Gdf], axis=1)
Gdf2 = Gdf2.fillna(np.nan)
Gdf2 = Gdf2.replace('', np.nan)
Gdf2 = pd.melt(Gdf2, id_vars=['Image_Name'], value_vars=['spk_0', 'spk_1', 'spk_2','spk_3'])
Gdf2 = Gdf2[['Image_Name', 'value']]
Gdf2 = Gdf2.dropna()
Gdf2 = Gdf2.rename(columns={"value": "Spike_Label"}, errors="raise")
Gdf2['Spike_Label'] = pd.to_numeric(Gdf2.Spike_Label, errors='coerce')
Gdf2['Green'] = 'Green'
Gdf2 = Gdf2.dropna()
Gdf2 = Gdf2.sort_values(by=['Image_Name'])
# Gdf2.to_csv('green.csv')
# Gdf2.dtypes
# Detect shed spikes
SList = TR_spk["Shed"].str.split(",", n = -1, expand = False)
Sdf = pd.DataFrame(SList)
Sdf = pd.DataFrame(Sdf.Shed.values.tolist(), Sdf.index).add_prefix('spk_')
Sdf2 = pd.concat([TR_spk[['Image_Name','Shed']], Sdf], axis=1)
Sdf2 = Sdf2.fillna(np.nan)
Sdf2 = Sdf2.replace('', np.nan)
Sdf2 = pd.melt(Sdf2, id_vars=['Image_Name'], value_vars=['spk_0', 'spk_1', 'spk_2'])
Sdf2 = Sdf2.dropna()
Sdf2 = Sdf2[['Image_Name', 'value']]
Sdf2 = Sdf2.rename(columns={"value": "Spike_Label"}, errors="raise")
Sdf2['Spike_Label'] = pd.to_numeric(Sdf2['Spike_Label'])
Sdf2['Shed'] = 'Shed'
Sdf2 = Sdf2.dropna()
Sdf2 = Sdf2.sort_values(by=['Image_Name'])
# Sdf2.to_csv('shed.csv')

# Detect false spikes
FList = TR_spk["FalseSpk"].str.split(",", n = -1, expand = False)
Fdf = pd.DataFrame(FList)
Fdf = pd.DataFrame(Fdf.FalseSpk.values.tolist(), Fdf.index).add_prefix('spk_')
Fdf2 = pd.concat([TR_spk[['Image_Name','FalseSpk']], Fdf], axis=1)
Fdf2 = Fdf2.fillna(np.nan)
Fdf2 = Fdf2.replace('', np.nan)
Fdf2 = pd.melt(Fdf2, id_vars=['Image_Name'], value_vars=['spk_0', 'spk_1','spk_2'])
Fdf2 = Fdf2.dropna()
Fdf2 = Fdf2[['Image_Name', 'value']]
Fdf2 = Fdf2.rename(columns={"value": "Spike_Label"}, errors="raise")
Fdf2['Spike_Label'] = pd.to_numeric(Fdf2['Spike_Label'], errors='coerce')
Fdf2['FalseSpk'] = 'FalseSpk'
Fdf2 = Fdf2.dropna()
Fdf2 = Fdf2.sort_values(by=['Image_Name'])
# Fdf2.to_csv('FalseSpk.csv')

# Detect envelopes
Edf = TR_spk[["Image_Name","Envelope"]]
Edf = Edf.rename(columns={"Envelope": "Spike_Label"}, errors="raise")
Edf['Envelope'] = 'Envelope'

# Merge datafiles
spk_df2 = pd.merge(Spikes_data, Gdf2, how='left', on = ['Image_Name','Spike_Label'])
spk_df2 = pd.merge(spk_df2, Sdf2, how='left', on = ['Image_Name','Spike_Label'])
spk_df2 = pd.merge(spk_df2, Fdf2, how='left', on = ['Image_Name','Spike_Label'])
spk_df2 = pd.merge(spk_df2, Edf, how='left', on = ['Image_Name','Spike_Label'])

# Subset bad "spikes"
spk_df3 = spk_df2.loc[spk_df2['Green'] != 'Green']
spk_df3 = spk_df3.loc[spk_df3['Shed'] != 'Shed']
spk_df3 = spk_df3.loc[spk_df3['FalseSpk'] != 'FalseSpk']
spk_df3 = spk_df3.loc[spk_df3['Envelope'] != 'Envelope']

# Create datasets with the exluded "spikes"
Green = spk_df2.loc[spk_df2['Green'] == 'Green']
Shed = spk_df2.loc[spk_df2['Shed'] == 'Shed']
FalseSpk = spk_df2.loc[spk_df2['FalseSpk'] == 'FalseSpk']
Envelope = spk_df2.loc[spk_df2['Envelope'] == 'Envelope']

```
